```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('/content/Parkinsson disease.csv')
```

```python
df = df.drop('name', axis=1)
```

```python
df.corr()
```

```python
plt.figure(figsize=(10, 7))
sns.heatmap(df.corr(),annot = True)
plt.show()
```

```python
factors = ['MDVP:Fo(Hz)',	'MDVP:Fhi(Hz)',	'MDVP:Flo(Hz)',	'MDVP:Jitter(%)',	'MDVP:Jitter(Abs)',	'MDVP:RAP',	'MDVP:PPQ',	'Jitter:DDP',	'MDVP:Shimmer',	'MDVP:Shimmer(dB)', 'Shimmer:APQ3','Shimmer:APQ5', 'MDVP:APQ',	'Shimmer:DDA',	'NHR',	'HNR', 'RPDE',	'DFA',	'spread1',	'spread2',	'D2',	'PPE']
y = df['status']
x = df[factors]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
model = RandomForestClassifier(n_estimators = 200, random_state = 0)
```

```python
model.fit(x_train,y_train)
```

```python
predictions = model.predict(x_test)
predictions
```

```python
accuracy = accuracy_score(y_test,predictions)
print(accuracy*100)
```

```python
confusion_matrix = confusion_matrix(y_test,predictions)
confusion_matrix
```

```python
fig, ax = plt.subplots(figsize=(7.5, 7.5))
ax.matshow(confusion_matrix, cmap=plt.cm.Blues, alpha=0.3)
for i in range(confusion_matrix.shape[0]):
    for j in range(confusion_matrix.shape[1]):
        ax.text(x=j, y=i,s=confusion_matrix[i, j], va='center', ha='center', size='xx-large')

plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.show()
```

```python
print(classification_report(y_test,predictions))
```

```python
feature_importance = model.feature_importances_
indices = np.argsort(feature_importance)
print("Feature Importance")
for i , idx in enumerate(indices):
  print(f"{i+1}. {feature_importance[idx]}: {feature_importance[idx]}")

plt.figure(figsize =(10,6))
plt.title("Feature Importances")
plt.bar(range(x.shape[1]), feature_importance[indices],align = "center")
plt.xticks(range(x.shape[1]),np.array(factors)[indices], rotation = 90)
plt.xlabel("Feature")
plt.ylabel("Importance")
plt.tight_layout()
plt.show()
```

```python
# prompt: how to remove outliers

def remove_outliers(df, col):
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
  return df

df = remove_outliers(df, 'MDVP:Jitter(%)')
df = remove_outliers(df, 'PPE')
```
